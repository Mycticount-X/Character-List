```python
import numpy as np

# TODO: buat function untuk cek diagonally dominant
def diagonal_dominant(mtx):
    # List -> Array
    mtx = np.array(mtx)
    
    # Ambil diagonal absolut
    diag = np.diag(np.abs(mtx))
    
    # Axis = 1 (Cari Row)
    other = np.sum(np.abs(mtx), axis=1) - diag
    
    # Pastikan semua elemen diagonal lebih besar dari jumlah elemen lain di barisnya
    if np.any(diag <= other):
        return False
    return True

# TODO: buat function gauss-seidel
def gauss_seidel(mtx, y, epsilon=0.01, max_iter=15):
    if not diagonal_dominant(mtx):
        print("Not Diagonally Dominant")
        return None
    
    x = np.array(mtx, dtype=float)
    y = np.array(y, dtype=float)
    
    diag = np.diag(x)
    
    x = -x
    np.fill_diagonal(x, 0)
    
    start = np.zeros_like(y)
    
    for k in range(max_iter):
        new = np.copy(start)
        
        for i, r in enumerate(x):
            new[i] = (y[i] + np.dot(r, start)) / diag[i]  # Gunakan start agar lebih stabil
        
        print(f'Iterasi {k}: {new}')
        
        dx = np.linalg.norm(new - start)  # Gunakan norm agar lebih optimal
        print(f'dx: {dx}')
        
        if dx < epsilon:
            print("Converged!")
            return new  # Kembalikan hasil
        
        start = new
    
    print("Not Converged")
    return None

# TODO: implement function gauss-seidel untuk persamaan yang lebih dari 1
Xm = [
    # Persamaan 1
    [
      [6, 3, 2],
      [2, 7, 3],
      [7, 5, 15]
    ],
    # Persamaan 2
    [
      [9, 2, 3, 2],
      [4, 11, 2, 3],
      [1, -4, 8, -1],
      [5, -3, -6, 16],
    ]
]

Ym = [
    # Hasil dari persamaan 1
    [8, 5, 10],
    # Hasil dari persamaan 2
    [15, 9, 15, 21]
]

for i, (x, y) in enumerate(zip(Xm, Ym)):
    print(f'Persamaan {i+1}:')
    result = gauss_seidel(x, y)
    if result is not None:
        print(f'Hasil akhir: {result}')
    print()
```
